```python
#this cell will create tables for each munici and contaminant.
# each table should contain around 30 days with its differents hours

stations = df['ESTACIO'].unique()
contaminants = df['CODI_CONTAMINANT'].unique()
tables = len(stations)*len(contaminants)

tables_list = []

for station in stations:
    for contaminant in contaminants:
        tables_list.append((df.loc[ (df['ESTACIO'] == station) & (df['CODI_CONTAMINANT'] == contaminant)]))

#this will print the first table:
tables_list[1]

```
